```python
import cv2
import os
import matplotlib.pyplot as plt

def image_negative(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Check if the image is read successfully
    if image is None:
        print("Error: Unable to read image.")
        return

    # Calculate the negative of the image
    negative_image = 255 - image

    # Display the original and negative images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
    plt.title('Negative Image')
    plt.axis('off')

    plt.show()

    # Save the negative image with ('negative'- prefix)
    directory = os.path.dirname(image_path)
    filename, extension = os.path.splitext(os.path.basename(image_path))
    negative_filename = f"negative_{filename}{extension}"
    negative_image_path = os.path.join(directory, negative_filename)
    cv2.imwrite(negative_image_path, negative_image)

    print(f"Negative image saved as: {negative_image_path}")
```

```python
# Paths for generated images (with random pixel intensities):
binary = 'GeneratedImages/binary.png'
grey = 'GeneratedImages/grey_image.png'
rgb = 'GeneratedImages/rgb_image.png'
```

### Image Negative for Binary Image: 

```python
image_negative(binary)
```

### Image Negative for GreyScale Image:

```python
image_negative(grey)
```

### Image Negative For RGB Image:

```python

image_negative(rgb)
```

### Image Negative For Real Images

```python
# Path to real images
image_path = 'Images/mountain.png' 

image_negative(image_path)
```
